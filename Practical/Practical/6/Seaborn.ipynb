```python
#Visualize the data using Python library Seaborn by plotting following graphs forest fire datasets 
#a. Scatter Plot 
#b. Box Plot 
#c. Histogram 
#d. Pie Chart 
#e. Line Chart 
```

```python
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns
```

```python
df=pd.read_csv('forestfires.csv')
df
```

```python
#a. Scatter Plot
sns.scatterplot(x='month', y='temp', data=df,hue='day')
plt.title('Scatter Chart')
plt.show()
```

```python
#b. Box Plot
df.boxplot(by ='day', column =['temp'], grid = False)
```

```python
#c. Histogram 
sns.histplot(df['DMC'])
plt.title("Histogram")
 # Adding the legends
plt.show()
```

```python
#d. Pie Chart
keys='FFMC','DMC','DC'
sizes=[df['FFMC'].sum(),df['DMC'].sum(),df['DC'].sum()]
 # Plotting data on chart
plt.pie(sizes, labels=keys, autopct='%.0f%%')
 # Add title to the chart
plt.title('Values of fire weather indices')
 # Displaying chart
plt.show()
```

```python
#e.Line chart
# draw lineplot
sns.lineplot(x="month", y="temp", data=df)
 # setting the title 
plt.title('Line Chart')
plt.show()
```

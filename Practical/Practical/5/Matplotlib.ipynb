```python
#Visualize the data using Python library matplotlib by plotting following graphs forest fire datasets 
#a. Scatter Plot 
#b. Bar Plot 
#c. Histogram 
#d. Pie Chart 
#e. Line Chart
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('forestfires.csv')
df
```

```python
 #a.Scatter Plot
 plt.scatter(df['month'], df['wind'])
 # Adding Title to the Plot
 plt.title("Scatter Plot")
 # Setting the X and Y labels
 plt.xlabel('Month')
 plt.ylabel('wind')
#plot
plt.show()
```

```python
#b.Bar Chart
 #Bar chart with day against temprature
plt.bar(df['day'], df['temp'])
plt.title("Bar Chart")
 # Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Temprature')
 # Adding the legends
plt.show()
```

```python
 #c.Histogram
 # histogram of total_bills
plt.hist(df['DMC'])
plt.title("Histogram")
 # Adding the legends
plt.show()
```

```python
#d.pie chart
keys='FFMC','DMC','DC'
sizes=[df['FFMC'].sum(),df['DMC'].sum(),df['DC'].sum()]
 # Plotting data on chart
plt.pie(sizes, labels=keys, autopct='%.0f%%')
 # Add title to the chart
plt.title('Values of fire weather indices')
 # Displaying chart
plt.show() 
```

```python
#e.Line chart
plt.plot(df['month'], df['temp'], marker='o', linestyle='-')
# Setting the title
plt.title('Line Chart')
# Labeling the axes
plt.xlabel('Month')
plt.ylabel('Temperature')
# Display the plot
plt.show()
```
